```python
import json

with open('ochuman_coco_format_copy.json', 'r') as data_file:
    data = json.load(data_file)

```

```python
#for img in data['images']:
for human_data in data['annotations']:
    
    if "segmentation" in human_data:
        human_data.pop('segmentation',None)
    if "num_keypoints" in human_data:
        human_data.pop('num_keypoints',None)
    if "keypoints" in human_data:
        human_data.pop('keypoints',None)
    if 'inner'in human_data:
        human_data.pop('inner',None)
    if 'links'in human_data:
        human_data.pop('links',None)
    if 'kpts'in human_data:
        human_data.pop('kpts',None)
    print (human_data)
 
```

```python
with open('ochuman_coco_format_copy.json', 'w') as data_file:
    data = json.dump(data, data_file)
```
